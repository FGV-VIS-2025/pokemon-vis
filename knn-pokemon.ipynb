```python
import pandas as pd
```

```python
pokemons = pd.read_csv("./data/pokemon.csv")
pokemons_species = pd.read_csv("./data/pokemon_species.csv")
pokemons_species_names = pd.read_csv("./data/pokemon_species_names.csv")
pokemons_species_names = pokemons_species_names[pokemons_species_names["local_language_id"] == 9]
pokemons_stats = pd.read_csv("./data/pokemon_stats_clean.csv")
pokemons_types = pd.read_csv("./data/pokemon_types.csv")
```

```python
species_ids_com_nome = pokemons_species_names['pokemon_species_id'].unique()
pokemons_validos = pokemons[pokemons['species_id'].isin(species_ids_com_nome)]
ids_validos = pokemons_validos['id'].unique()
pokemons_stats = pokemons_stats[pokemons_stats['Pokemon_Id'].isin(ids_validos)]
pokemons_stats.shape
```

```python
pokemons_stats.shape
```

```python
id_to_type = {
    1: "Normal",
    2: "Lutador",
    3: "Voador",
    4: "Venenoso",
    5: "Terrestre",
    6: "Pedra",
    7: "Inseto",
    8: "Fantasma",
    9: "Aço",
    10: "Fogo",
    11: "Água",
    12: "Grama",
    13: "Elétrico",
    14: "Psíquico",
    15: "Gelo",
    16: "Dragão",
    17: "Sombrio",
    18: "Fada",
    19: "Estelar",
    10001: "Desconhecido",
    10002: "Sombra"
}
pokemons_types['type_name'] = pokemons_types['type_id'].map(id_to_type)

# Reorganiza aux_df para ter uma linha por pokémon com colunas type_1 e type_2
types_pivot = pokemons_types.pivot(index='pokemon_id', columns='slot', values='type_name')
types_pivot.columns = ['type_1', 'type_2']
types_pivot = types_pivot.reset_index()


types_pivot.head()
```

```python
df_base = pd.merge(pokemons_species, pokemons_species_names, left_on='id', right_on='pokemon_species_id', how='inner')
print("Após merge species + names:", df_base.shape)

df_poke = pd.merge(pokemons, df_base, left_on='species_id', right_on='id', how='inner')
print("Após merge pokemons + base:", df_poke.shape)

df_stats = pd.merge(df_poke, pokemons_stats, left_on='id_x', right_on='Pokemon_Id', how='inner')
print("Após merge com stats:", df_stats.shape)

aux_df = pd.merge(df_stats, types_pivot, left_on='id_x', right_on='pokemon_id', how='inner')
print("Após merge com tipos:", aux_df.shape)
```

```python
id_to_shape = {
    1: "Esférico",
    2: "Rabiscado",
    3: "Peixe",
    4: "Com braços",
    5: "Massa informe",
    6: "Ereto",
    7: "Com pernas",
    8: "Quadrúpede",
    9: "Com asas",
    10: "Com tentáculos",
    11: "Com cabeças múltiplas",
    12: "Humanoide",
    13: "Com asas de inseto",
    14: "Com armadura"
}

aux_df["shape"] = aux_df["shape_id"].map(id_to_shape)
```

```python
aux_df.columns
```

```python
colunas = ["id_x", "height", "weight", "base_happiness", "is_baby", 
           "is_legendary", "is_mythical", "genus", "Hp_Stat", "Attack_Stat", 
           "Defense_Stat", "Special_Attack_Stat", "Special_Defense_Stat", 
           "Speed_Stat", "type_1", "type_2", "shape",]
```

```python
final_df = aux_df[colunas]
final_df.columns = ["id", "height", "weight", "base_happiness", "is_baby", 
                     "is_legendary", "is_mythical", "genus", "hp_stat", 
                     "attack_stat", "defense_stat", "special_attack_stat", 
                     "special_defense_stat", "speed_stat", "type_1", 
                     "type_2", "shape"]
final_df.head()
```

```python
import os

caminho_das_imagens = "../pokemon-vis/assets/pokemons"

# Lista para armazenar os IDs válidos
imagens_ids = []

# Itera sobre os arquivos na pasta
for nome in os.listdir(caminho_das_imagens):
    if nome.endswith(".png"):
        try:
            imagens_ids.append(int(nome.replace(".png", "")))
        except ValueError:
            pass  # ignora arquivos que não seguem o formato numérico

# Filtra o DataFrame
final_df = final_df[final_df["id"].isin(imagens_ids)]
print(final_df.shape)
```

```python
# normalizando variaveis numéricas
for col in ["height", "weight", "base_happiness", "hp_stat", 
            "attack_stat", "defense_stat", "special_attack_stat", 
            "special_defense_stat", "speed_stat"]:
    final_df[col] = round((final_df[col] - final_df[col].mean()) / final_df[col].std(), 3)
```

```python
final_df.to_csv("knn_data.csv", index=False)
```
