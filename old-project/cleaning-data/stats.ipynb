```python
import pandas as pd
```

```python
pokemon_stats = pd.read_csv("../../data/pokemon_stats.csv")
stats_name = pd.read_csv("../../data/stat_names.csv")
stats_name = stats_name[stats_name["local_language_id"] == 9]
```

```python
pokemon_stats.head()
```

```python
stats_name.head()
```

```python
stats_name_dict = {}
for each_row in range(stats_name.shape[0]):
    stat_id = int(stats_name.iloc[each_row]["stat_id"])
    stat_name = stats_name.iloc[each_row]["name"].title()
    stats_name_dict[stat_id] = stat_name
```

```python
stats_name_dict
```

```python
import pandas as pd

relevant_stats = {k: v for k, v in stats_name_dict.items()} 

columns = ["Pokemon_Id"]
for stat in relevant_stats.values():
    columns.extend([f"{stat}_Stat", f"{stat}_Effort"])

final_df = pd.DataFrame(columns=columns)

for pokemon_id, group in pokemon_stats.groupby("pokemon_id"):
    row = {"Pokemon_Id": pokemon_id}
    for _, stat_row in group.iterrows():
        stat_id = stat_row["stat_id"]
        if stat_id in relevant_stats:
            stat_name = relevant_stats[stat_id]
            row[f"{stat_name}_Stat"] = stat_row["base_stat"]
            row[f"{stat_name}_Effort"] = stat_row["effort"]
    final_df = pd.concat([final_df, pd.DataFrame([row])], ignore_index=True)
```

```python
final_df.to_csv("../../data/pokemon_stats_clean.csv")
```
